```python
def allocate_products():
    # Take user input for customers
    customers = input("Enter customers (comma-separated): ").strip().split(',')
    customers = [int(customer.strip()) for customer in customers]

    # Initialize dictionaries for demand and service
    customer_demand = {customer: 0 for customer in customers}
    customer_service = {customer: 0 for customer in customers}

    # Take user input for demand for each customer
    for customer in customers:
        valid_input = False
        while not valid_input:
            try:
                demand = int(input(f"Enter demand for customer {customer}: "))
                if demand >= 0:
                    customer_demand[customer] = demand
                    valid_input = True
                else:
                    print("Demand must be a non-negative integer.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")

    # Take user input for allocation limit
    valid_limit = False
    while not valid_limit:
        try:
            limit = int(input("Enter allocation limit: "))
            if limit >= 0:
                valid_limit = True
            else:
                print("Limit must be a non-negative integer.")
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

    rounds = 0
    total_demand = sum(customer_demand.values())
    total_allocated = 0

    print(f"\nInitial Total Demand: {total_demand} units\n")

    while total_allocated < limit and total_demand > 0:
        rounds += 1
        print(f"Round {rounds}:")

        for customer in customers:
            if customer_demand[customer] > 0 and total_allocated < limit:
                if customer_service[customer] < customer_demand[customer]:  # Check if demand is not fully met
                    print(f"Allocating 1 unit to customer {customer}")
                    customer_service[customer] += 1
                    total_allocated += 1
                    total_demand -= 1

            if total_allocated >= limit or total_demand == 0:
                break

        print("")

    # Calculate and display results
    for customer in customers:
        actual_demand = customer_demand[customer]
        serviced_units = customer_service[customer]

        if actual_demand > 0:
            service_percentage = (serviced_units / actual_demand) * 100
        else:
            service_percentage = 100.0  # Customer got all demanded units

        print(f"Customer {customer}:")
        print(f"  Demand: {actual_demand} units")  # Display original demand entered by user
        print(f"  Service: {serviced_units} units")
        print(f"  Service Percentage: {service_percentage:.2f}%")
        print("")

# Example usage:
allocate_products()
```
